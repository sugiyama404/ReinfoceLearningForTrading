```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def act(epsilon,num,r = 0.995):
    epsilon_min = 0.01
    if np.random.rand() <= epsilon:
        num += 1
    if epsilon > epsilon_min:
        epsilon *= r
    return epsilon,num
```

```python
#データフレームの作成.
df = pd.DataFrame(index=[], columns=["epsilon", "num"])
df2 = pd.DataFrame(index=[], columns=["epsilon", "num"])
df3 = pd.DataFrame(index=[], columns=["epsilon", "num"])

epsilon = 1.0
num = 0
r = 0.995

for i in range(1000):
    epsilon,num = act(epsilon,num,r = r)
    tmp_se = pd.Series( [epsilon,num], index=df.columns, name = i )
    df = df.append( tmp_se )

epsilon = 1.0
r = 0.5
for i in range(1000):
    epsilon,num = act(epsilon,num,r = r)
    tmp_se = pd.Series( [epsilon,num], index=df2.columns, name = i )
    df2 = df2.append( tmp_se )

epsilon = 1.0
r = 0.1
for i in range(1000):
    epsilon,num = act(epsilon,num,r = r)
    tmp_se = pd.Series( [epsilon,num], index=df3.columns, name = i )
    df3 = df3.append( tmp_se )

df.plot(secondary_y=['num']).set_title('r = 0.995')
df2.plot(secondary_y=['num']).set_title('r = 0.5')
df3.plot(secondary_y=['num']).set_title('r = 0.1')
plt.show()
```
